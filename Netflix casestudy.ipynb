## Data Analysis

```python
df.isnull().sum()
```

```python
# Handling the missing values
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna('unknown')
df['duration'] = df['duration'].fillna('unknown')
```

## Exploratory Data Analysis

```python
#Number of Titles released per year 
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,5))
sns.countplot(x='release_year', data=df)
plt.title("Number of Titles Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

## Key Insights
- From the above graph we got to know that most of the titles are released in the year 2018 whose count is more than 1000.

```python
#distribution of Movies vs TV Shows
types = df['type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(types,labels=types.index,autopct="%0.1f%%")
plt.title("Distribution of Movies vs TV shows")
plt.show()
```

## Key Insights
- Around 69.6% of the titles are Movie type whereas 30.4% are of TV Shows

```python
df_directors = df[df['director'] != 'Unknown']
top_directors = df_directors['director'].value_counts().head(10)
```

```python
print(top_directors)
```

```python
#Top 10 directors and their movie count
plt.figure(figsize=(6,6))
plt.pie(top_directors,labels=top_directors.index,autopct="%0.1f%%")
plt.title("Top 10 Directors")
plt.show()
```

## Key Insight
- Among the top 10 directors, we can see RAJIV CHILAKA movies are around 13.3% securing the highest movie maker

## Key Insight
